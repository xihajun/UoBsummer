```python
def print_ast(cursor, deep=0):
    print(' '.join((deep*'    ', str(cursor.kind), str(cursor.spelling), str(cursor.location.file))))
    for child in cursor.get_children():
        print_ast(child,  deep+1)

print_ast(root)
```

```python
def print_small_ast(cursor, deep=0):
    if str(cursor.location.file) == juliet.iloc[0].filename:
        print(' '.join((deep*'    ', str(cursor.kind), str(cursor.spelling), str(cursor.location.file))))
    for child in cursor.get_children():
        print_small_ast(child,  deep+1)

print_small_ast(root)
```
